```python
from langchain.document_loaders import UnstructuredURLLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain.vectorstores import Pinecone
import pinecone
from langchain.embeddings import HuggingFaceEmbeddings
from transformers import AutoTokenizer, AutoModelForCausalLM
import transformers
from langchain import HuggingFacePipeline
from huggingface_hub import notebook_login
import torch
```

```python
URLs = [
    "https://www.mlvti.ac.in/"
]
```

```python
loader = UnstructuredURLLoader(urls = URLs)
data = loader.load()
```

```python
data
```

```python
text_splitter = CharacterTextSplitter(separator='\n',
                      chunk_size = 1000,
                      chunk_overlap = 200)
```

```python
text_chunks = text_splitter.split_documents(data)
```

```python
text_chunks
```

```python
len(text_chunks)
```

# Embedding model

```python
embeddings = HuggingFaceEmbeddings(model_name= "sentence-transformers/all-MiniLM-L6-v2")
```

```python
import pinecone
# initialize pinecone
pinecone.init(
    api_key = PINECONE_API_KEY, #find at app.pinecone.io
    environment=PINECONE_API_ENV # next to api key in console
)
index_name = "mlvt-bot" # put in the name of your pinecone index here
```

```python
docsearch = Pinecone.from_texts([t.page_content for t in text_chunks], embeddings, index_name=index_name)
```

# Create LLM wrapper

```python
notebook_login()
```

```python
tokenizer = AutoTokenizer.from_pretrained("meta-llama/Llama-2-7b-chat-hf",
                                          use_auth_token=True,)
```

```python
model = AutoModelForCausalLM.from_pretrained("meta-llama/Llama-2-7b-chat-hf",
                                             device_map='auto',
                                             torch_dtype=torch.float16,
                                             use_auth_token=True,
                                             load_in_8bit=True,
                                             #load_in_4bit=True
                                             )
```

```python
pipe = transformers.pipeline("text-generation",
                             model=model,
                             tokenizer = tokenizer,
                             torch_dtype = torch.bfloat16,
                             device_map="auto",
                             max_new_tokens = 512,
                             do_sample = True,
                             top_k = 30,
                             num_return_sequences=1,
                             eos_token_id=tokenizer.eos_token_id
                             )
```

```python
llm = HuggingFacePipeline(pipeline = pipe, model_kwargs={'temprature': 0})
```

```python
llm.predict("Please provide a consise summary of Indian culture")
```

# Initialize the Retrieval QA

```python
from langchain.chains import RetrievalQA
```

```python
qa = RetrievalQA.from_chain_type(llm = llm, chain_type = "stuff", retriever=docsearch.as_retriever())
```

```python
query = "what types of events happens in this college"
```

```python
print(qa.run(query))
```
